##### Funciones de activaci√≥n

```python
def sigmoide(a):
    return 1 / (1 + np.exp(-a))
```

```python
def step(a):
    return np.piecewise(a, [a<0.0, a>0.0], [0, 1])
```

```python
def relu(a):
    return np.maximum(0,a)
```

```python
def tanh(a):
    return np.tanh(a)
```

##### Funciones de p√©rdida

```python
def mse(y, y_hat, derivate=False): # Mean Square Error
    if derivate:
        return (y_hat - y)
    return np.mean((y_hat - y) ** 2)
```

```python
real = np.array([0,0,1,1])
prediccion = np.array([0.9, 0.5, 0.2, 0.0])
```

```python
mse(real, prediccion)
```
